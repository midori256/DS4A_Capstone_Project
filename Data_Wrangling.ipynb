```python
path_list = [
    "data/CA_Any Opioid-Related Overdose_Death_by County_2015_12.08.2021.csv",
    "data/CA_Any Opioid-Related Overdose_Death_by County_2016_12.08.2021.csv",
    "data/CA_Any Opioid-Related Overdose_Death_by County_2017_12.08.2021.csv",
    "data/CA_Any Opioid-Related Overdose_Death_by County_2018_12.08.2021.csv",
    "data/CA_Any Opioid-Related Overdose_Death_by County_2019_12.08.2021.csv",
    "data/CA_Any Opioid-Related Overdose_Death_by County_2020_12.08.2021.csv"]

data_2015 = pd.read_csv(path_list[0], skiprows=2)
data_2016 = pd.read_csv(path_list[1], skiprows=2)
data_2017 = pd.read_csv(path_list[2], skiprows=2)
data_2018 = pd.read_csv(path_list[3], skiprows=2)
data_2019 = pd.read_csv(path_list[4], skiprows=2)
data_2020 = pd.read_csv(path_list[5], skiprows=2)
```

```python
# Keep only Rates and Counts columns
dfs = [data_2015,data_2016,data_2017,data_2018,data_2019,data_2020]
for item in dfs:
    item.drop(columns=["95% LCL","95% UCL"], inplace=True)
```

```python
# Add Year column into each dataframe before concatenate
for index,value in enumerate(dfs):
    value["Year"] = 2015 + index
```

```python
# Concatenate all year into one dataframe
all_df = pd.concat(dfs)
all_df.head()
```

Now we have a dataframe that includes number of opioid-related deaths in Counts, and age-adjusted rates in Rates, for each California County, from 2015 to 2020
